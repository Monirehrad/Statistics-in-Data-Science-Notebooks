# Chapter 7: Confidence Intervals and Their Meaning


In this notebook, we explore confidence intervals and how they measure uncertainty.

We will study:
- point estimates
- margin of error
- confidence levels
- confidence intervals for means
- confidence intervals for proportions
- one-sided confidence intervals


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Point Estimates

```python
# Sample average waiting time in minutes

sample_mean = 18.4

print("Point estimate:", sample_mean)
```

## From a Point Estimate to an Interval

```python
# Margin of error example

margin_error = 1.6

lower = sample_mean - margin_error
upper = sample_mean + margin_error

print("Confidence interval:")
print(round(lower, 1), "to", round(upper, 1))
```

## Standard Error

```python
# Battery lifetime example

sample_std = 40
sample_size = 25

standard_error = sample_std / np.sqrt(sample_size)

print("Standard error:", standard_error)
```

## Confidence Level

```python
# Critical value for a 95% confidence interval

critical_value = stats.t.ppf(0.975, df=24)

print("Critical value:", round(critical_value, 3))
```

## Visualizing a 95% Confidence Interval

```python
# Create values for the horizontal axis

x = np.linspace(-4, 4, 300)

# Standard normal density

y = stats.norm.pdf(x)

plt.plot(x, y)

plt.axvline(-1.96, linestyle='--')
plt.axvline(1.96, linestyle='--')

plt.title("Middle 95% of the Distribution")

plt.xlabel("z-score")
plt.ylabel("Density")

plt.show()
```

## A Complete Confidence Interval Example

```python
# Battery lifetime example

sample_mean = 1012
sample_std = 40
sample_size = 25
```

```python
# Compute the standard error

standard_error = sample_std / np.sqrt(sample_size)

print("Standard error:", standard_error)
```

```python
# Critical value for a 95% confidence interval

critical_value = stats.t.ppf(0.975, df=24)

print("Critical value:", round(critical_value, 3))
```

```python
# Compute the margin of error

margin_error = critical_value * standard_error

print("Margin of error:", round(margin_error, 2))
```

```python
# Construct the confidence interval

lower = sample_mean - margin_error
upper = sample_mean + margin_error

print("95% Confidence Interval:")
print(round(lower, 2), "to", round(upper, 2))
```

## Confidence Interval for a Proportion

```python
# 120 out of 200 customers prefer a new product

successes = 120
sample_size = 200

sample_proportion = successes / sample_size

print("Sample proportion:", sample_proportion)
```

```python
# Standard error for proportions

standard_error = np.sqrt(
    (sample_proportion * (1 - sample_proportion)) / sample_size
)

print("Standard error:", round(standard_error, 4))
```

```python
# Critical value for a 95% confidence interval

z_star = 1.96
```

```python
# Construct the confidence interval

margin_error = z_star * standard_error

lower = sample_proportion - margin_error
upper = sample_proportion + margin_error

print("95% Confidence Interval:")
print(round(lower, 3), "to", round(upper, 3))
```

## Confidence Intervals and Hypothesis Testing

```python
# Example confidence interval

lower = 103
upper = 108

print("95% Confidence Interval:")
print(lower, "to", upper)

print()
print("Hypothesized value: 100")
```

```python
# Null hypothesis:
# H0: mu = 100

# Alternative hypothesis:
# Ha: mu != 100

print("H0: mu = 100")
print("Ha: mu != 100")
```

```python
# Check whether the hypothesized value lies inside the interval

hypothesized_value = 100

if lower <= hypothesized_value <= upper:
    print("Fail to reject H0")
else:
    print("Reject H0")
```

# Appendix: One-Sided Confidence Intervals

## Upper One-Sided Confidence Interval

```python
# Sample information

sample_mean = 82
sample_std = 10
sample_size = 50
```

```python
# Critical value for a one-sided 95% confidence interval

z_star = 1.645

print("Critical value:", z_star)
```

```python
# Compute the standard error

standard_error = sample_std / np.sqrt(sample_size)

print("Standard error:", round(standard_error, 2))
```

```python
# Construct the upper confidence bound

upper_bound = sample_mean + z_star * standard_error

print("Upper 95% confidence bound:")
print(round(upper_bound, 2))
```

## Lower One-Sided Confidence Interval

```python
# Construct the lower confidence bound

lower_bound = sample_mean - z_star * standard_error

print("Lower 95% confidence bound:")
print(round(lower_bound, 2))
```

## Comparing Two-Sided and One-Sided Critical Values

```python
# Two-sided and one-sided critical values

two_sided = 1.96
one_sided = 1.645

print("Two-sided critical value:", two_sided)
print("One-sided critical value:", one_sided)
```

## Visualizing a One-Sided Confidence Interval

```python
# Create values for the horizontal axis

x = np.linspace(-4, 4, 300)

# Standard normal density

y = stats.norm.pdf(x)

plt.plot(x, y)

# One-sided critical value

plt.axvline(1.645, linestyle='--')

plt.title("One-Sided 95% Confidence Interval")

plt.xlabel("z-score")
plt.ylabel("Density")

plt.show()
```
