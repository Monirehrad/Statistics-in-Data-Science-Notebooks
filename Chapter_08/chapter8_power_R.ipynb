## Type I Error

```r
# Null hypothesis:
# H0: the patient is healthy

# Alternative hypothesis:
# Ha: the patient has the disease

cat("H0: patient is healthy\n")
cat("Ha: patient has the disease\n")
```

```r
# Type I error:
# Rejecting H0 even though H0 is true

cat("Type I Error = False Positive\n")
```

## Type II Error

```r
# Null hypothesis:
# H0: the patient is healthy

# Alternative hypothesis:
# Ha: the patient has the disease

cat("H0: patient is healthy\n")
cat("Ha: patient has the disease\n")
```

```r
# Type II error:
# Failing to reject H0 even though H0 is false

cat("Type II Error = False Negative\n")
```

## Balancing the Two Errors

```r
alpha_1 <- 0.05
alpha_2 <- 0.01

cat("First significance level:", alpha_1, "\n")
cat("Second significance level:", alpha_2, "\n")
```

```r
cat("Reducing alpha makes rejection harder.\n")
```

## Visualizing Type I and Type II Errors

```r
x <- seq(-4, 8, length.out = 400)

null_distribution <- dnorm(x, mean = 0, sd = 1)
alternative_distribution <- dnorm(x, mean = 2.5, sd = 1)

critical_value <- 2
```

```r
plot(x, null_distribution, type="l",
     xlab="Test Statistic", ylab="Density",
     main="Type I and Type II Errors")
lines(x, alternative_distribution, lty=2)
abline(v=critical_value, lty=3)
legend("topright",
       legend=c("Null Distribution","Alternative Distribution","Critical Value"),
       lty=c(1,2,3))
```

## Statistical Power

```r
beta <- 0.20
power <- 1 - beta
cat("Statistical power:", power, "\n")
```

## Effect of Sample Size

```r
sigma <- 12
n_small <- 25
n_large <- 100
```

```r
se_small <- sigma / sqrt(n_small)
cat("Standard error (small sample):", se_small, "\n")
```

```r
se_large <- sigma / sqrt(n_large)
cat("Standard error (large sample):", se_large, "\n")
```

```r
cat("Larger samples increase statistical power.\n")
```

## Effect Size

```r
conversion_old <- 0.05
conversion_new <- 0.052
small_effect <- conversion_new-conversion_old
cat("Small effect size:", small_effect, "\n")
```

```r
conversion_old <- 0.05
conversion_new <- 0.15
large_effect <- conversion_new-conversion_old
cat("Large effect size:", large_effect, "\n")
```

```r
cat("Larger effect sizes increase statistical power.\n")
```

## Statistical Significance vs Practical Significance

```r
old_average <- 30.0
new_average <- 30.2
difference <- new_average-old_average
cat("Observed difference:", difference, "\n")
```

```r
cat("Always consider practical importance.\n")
```

# Appendix D: Statistical Power and Sample Size

## Why Sample Size Matters

```r
sigma <- 20
n_1 <- 25
n_2 <- 100
se_1 <- sigma/sqrt(n_1)
se_2 <- sigma/sqrt(n_2)
cat("Standard error with n = 25:", se_1, "\n")
cat("Standard error with n = 100:", se_2, "\n")
```

```r
cat("Larger samples produce more stable estimates.\n")
```

## Effect Size and Distribution Overlap

```r
x <- seq(-5,8,length.out=400)
small_effect_distribution <- dnorm(x, mean=1, sd=1)
large_effect_distribution <- dnorm(x, mean=3, sd=1)
```

```r
plot(x, small_effect_distribution, type="l",
main="Effect Size and Distribution Separation",
xlab="Test Statistic", ylab="Density")
lines(x, large_effect_distribution, lty=2)
legend("topright", legend=c("Small Effect","Large Effect"), lty=c(1,2))
```

## A Simple Sample Size Calculation

```r
alpha <- 0.05
power <- 0.80
```

```r
z_alpha <- qnorm(0.975)
z_beta <- qnorm(power)

cat("z_alpha:", round(z_alpha,2), "\n")
cat("z_beta:", round(z_beta,2), "\n")
```

```r
sigma <- 12
delta <- 5
```

```r
n <- ((z_alpha + z_beta) * sigma / delta)^2
cat("Required sample size:", round(n,1), "\n")
```

```r
required_sample_size <- ceiling(n)
cat("Recommended sample size:", required_sample_size, "\n")
```
