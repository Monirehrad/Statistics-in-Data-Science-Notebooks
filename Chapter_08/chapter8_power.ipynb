# Chapter 8: Errors, Power, and Real-World Tradeoffs


In this notebook, we explore statistical errors, statistical power,
and the practical tradeoffs involved in statistical decision-making.

We will study:
- Type I and Type II errors
- significance levels
- statistical power
- practical vs statistical significance
- sample size and power


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Type I Error

```python
# Null hypothesis:
# H0: the patient is healthy

# Alternative hypothesis:
# Ha: the patient has the disease

print("H0: patient is healthy")
print("Ha: patient has the disease")
```

```python
# Type I error:
# Rejecting H0 even though H0 is true

print("Type I Error = False Positive")
```

## Type II Error

```python
# Null hypothesis:
# H0: the patient is healthy

# Alternative hypothesis:
# Ha: the patient has the disease

print("H0: patient is healthy")
print("Ha: patient has the disease")
```

```python
# Type II error:
# Failing to reject H0 even though H0 is false

print("Type II Error = False Negative")
```

## Balancing the Two Errors

```python
# Compare two significance levels

alpha_1 = 0.05
alpha_2 = 0.01

print("First significance level:", alpha_1)
print("Second significance level:", alpha_2)
```

```python
# Smaller significance levels reduce Type I errors
# but increase the chance of Type II errors.

print("Reducing alpha makes rejection harder.")
```

## Visualizing Type I and Type II Errors

```python
# Create values for the horizontal axis

x = np.linspace(-4, 8, 400)

# Null distribution

null_distribution = stats.norm.pdf(x, loc=0, scale=1)

# Alternative distribution

alternative_distribution = stats.norm.pdf(x, loc=2.5, scale=1)

# Critical value

critical_value = 2
```

```python
# Plot the two distributions

plt.plot(x, null_distribution, label='Null Distribution')

plt.plot(x, alternative_distribution,
         linestyle='--',
         label='Alternative Distribution')

plt.axvline(critical_value,
            linestyle=':',
            label='Critical Value')

plt.title("Type I and Type II Errors")

plt.xlabel("Test Statistic")
plt.ylabel("Density")

plt.legend()

plt.show()
```

## Statistical Power

```python
# Statistical power measures the probability
# of correctly rejecting a false null hypothesis.

beta = 0.20

power = 1 - beta

print("Statistical power:", power)
```

## Effect of Sample Size

```python
# Standard deviation

sigma = 12

# Compare two sample sizes

n_small = 25
n_large = 100
```

```python
# Standard error for the smaller sample

se_small = sigma / np.sqrt(n_small)

print("Standard error (small sample):", se_small)
```

```python
# Standard error for the larger sample

se_large = sigma / np.sqrt(n_large)

print("Standard error (large sample):", se_large)
```

```python
# Larger samples reduce the standard error
# and make real effects easier to detect.

print("Larger samples increase statistical power.")
```

## Effect Size

```python
# Small effect example

conversion_old = 0.05
conversion_new = 0.052

small_effect = conversion_new - conversion_old

print("Small effect size:", small_effect)
```

```python
# Large effect example

conversion_old = 0.05
conversion_new = 0.15

large_effect = conversion_new - conversion_old

print("Large effect size:", large_effect)
```

```python
# Larger effects are easier to detect statistically.

print("Larger effect sizes increase statistical power.")
```

## Statistical Significance vs Practical Significance

```python
# Website redesign example

old_average = 30.0
new_average = 30.2

difference = new_average - old_average

print("Observed difference:", difference)
```

```python
# A result can be statistically significant
# but still too small to matter in practice.

print("Always consider practical importance.")
```

# Appendix D: Statistical Power and Sample Size

## Why Sample Size Matters

```python
# Standard error formula

sigma = 20

n_1 = 25
n_2 = 100

se_1 = sigma / np.sqrt(n_1)
se_2 = sigma / np.sqrt(n_2)

print("Standard error with n = 25:", se_1)
print("Standard error with n = 100:", se_2)
```

```python
# Larger sample sizes reduce sampling variability.

print("Larger samples produce more stable estimates.")
```

## Effect Size and Distribution Overlap

```python
# Create values for the horizontal axis

x = np.linspace(-5, 8, 400)

# Small effect

small_effect_distribution = stats.norm.pdf(x, loc=1, scale=1)

# Large effect

large_effect_distribution = stats.norm.pdf(x, loc=3, scale=1)
```

```python
# Visualize the distributions

plt.plot(x, small_effect_distribution,
         label='Small Effect')

plt.plot(x, large_effect_distribution,
         linestyle='--',
         label='Large Effect')

plt.title("Effect Size and Distribution Separation")

plt.xlabel("Test Statistic")
plt.ylabel("Density")

plt.legend()

plt.show()
```

## A Simple Sample Size Calculation

```python
# Desired significance level and power

alpha = 0.05
power = 0.80
```

```python
# Critical values from the standard normal distribution

z_alpha = stats.norm.ppf(0.975)

z_beta = stats.norm.ppf(power)

print("z_alpha:", round(z_alpha, 2))
print("z_beta:", round(z_beta, 2))
```

```python
# Example values

sigma = 12
delta = 5
```

```python
# Sample size calculation

n = ((z_alpha + z_beta) * sigma / delta) ** 2

print("Required sample size:", round(n, 1))
```

```python
# Round up to the next whole number

required_sample_size = int(np.ceil(n))

print("Recommended sample size:", required_sample_size)
```
