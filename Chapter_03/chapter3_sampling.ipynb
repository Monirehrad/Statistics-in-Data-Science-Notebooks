# Chapter 3: Sampling and the Logic of Inference


In this notebook, we explore how samples are used to learn about populations.

We will study:
- populations and samples
- sampling variability
- sampling distributions
- standard error
- the Central Limit Theorem
- confidence intervals


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Population and Sample

```python
# Generate a population

population = np.random.normal(loc=70, scale=10, size=10000)
```

```python
# Visualize the population

plt.hist(population, bins=30)

plt.title("Population Distribution")

plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

```python
# Take a random sample from the population

sample = np.random.choice(population, size=50)
```

```python
# Compare the population mean and sample mean

print("Population mean:", round(np.mean(population), 2))
print("Sample mean:", round(np.mean(sample), 2))
```

## Sampling Variability

```python
# Generate three different random samples

sample_1 = np.random.choice(population, size=50)
sample_2 = np.random.choice(population, size=50)
sample_3 = np.random.choice(population, size=50)
```

```python
# Compare the sample means

print("Sample 1 mean:", round(np.mean(sample_1), 2))
print("Sample 2 mean:", round(np.mean(sample_2), 2))
print("Sample 3 mean:", round(np.mean(sample_3), 2))
```

## Sampling Distribution

```python
# Store sample means

sample_means = []
```

```python
# Generate many sample means

for i in range(1000):

    sample = np.random.choice(population, size=30)

    sample_mean = np.mean(sample)

    sample_means.append(sample_mean)
```

```python
# Visualize the sampling distribution

plt.hist(sample_means, bins=30)

plt.title("Sampling Distribution of the Sample Mean")

plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.show()
```

## Standard Error

```python
# Calculate the standard error

population_std = np.std(population)

sample_size = 30

standard_error = population_std / np.sqrt(sample_size)

print("Standard error:", round(standard_error, 2))
```

## Central Limit Theorem

```python
# Generate a skewed population

skewed_population = np.random.exponential(scale=2, size=10000)
```

```python
# Visualize the skewed population

plt.hist(skewed_population, bins=30)

plt.title("Skewed Population")

plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

```python
# Store sample means from the skewed population

clt_means = []
```

```python
# Generate many sample means

for i in range(1000):

    sample = np.random.choice(skewed_population, size=30)

    sample_mean = np.mean(sample)

    clt_means.append(sample_mean)
```

```python
# Visualize the sampling distribution

plt.hist(clt_means, bins=30)

plt.title("Sampling Distribution of Sample Means")

plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.show()
```

## Confidence Intervals

```python
# Take a random sample

sample = np.random.choice(population, size=50)
```

```python
# Calculate the sample mean

sample_mean = np.mean(sample)

print("Sample mean:", round(sample_mean, 2))
```

```python
# Calculate the margin of error

sample_std = np.std(sample)

margin_error = 1.96 * (sample_std / np.sqrt(len(sample)))
```

```python
# Calculate the confidence interval

lower = sample_mean - margin_error
upper = sample_mean + margin_error

print("95% Confidence Interval:")
print(round(lower, 2), "to", round(upper, 2))
```
