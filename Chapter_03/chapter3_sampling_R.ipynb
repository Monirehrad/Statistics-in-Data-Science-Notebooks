In this notebook, we explore how samples are used to learn about populations.

We will study:
- populations and samples
- sampling variability
- sampling distributions
- standard error
- the Central Limit Theorem
- confidence intervals


## Population and Sample

```r
# Generate a population

set.seed(42)

population <- rnorm(n = 10000, mean = 70, sd = 10)
```

```r
# Visualize the population

hist(
  population,
  breaks = 30,
  main = "Population Distribution",
  xlab = "Value",
  ylab = "Frequency",
  col = "lightblue"
)
```

```r
# Take a random sample from the population

sample_values <- sample(population, size = 50)
```

```r
# Compare the population mean and sample mean

cat("Population mean:", round(mean(population), 2), "\n")
cat("Sample mean:", round(mean(sample_values), 2), "\n")
```

## Sampling Variability

```r
# Generate three different random samples

sample_1 <- sample(population, size = 50)
sample_2 <- sample(population, size = 50)
sample_3 <- sample(population, size = 50)
```

```r
# Compare the sample means

cat("Sample 1 mean:", round(mean(sample_1), 2), "\n")
cat("Sample 2 mean:", round(mean(sample_2), 2), "\n")
cat("Sample 3 mean:", round(mean(sample_3), 2), "\n")
```

## Sampling Distribution

```r
# Store sample means

sample_means <- c()
```

```r
# Generate many sample means

for (i in 1:1000) {
  sample_values <- sample(population, size = 30)
  sample_mean <- mean(sample_values)
  sample_means <- c(sample_means, sample_mean)
}
```

```r
# Visualize the sampling distribution

hist(
  sample_means,
  breaks = 30,
  main = "Sampling Distribution of the Sample Mean",
  xlab = "Sample Mean",
  ylab = "Frequency",
  col = "lightblue"
)
```

## Standard Error

```r
# Calculate the standard error

population_std <- sd(population)

sample_size <- 30

standard_error <- population_std / sqrt(sample_size)

cat("Standard error:", round(standard_error, 2), "\n")
```

## Central Limit Theorem

```r
# Generate a skewed population

set.seed(42)

skewed_population <- rexp(n = 10000, rate = 1 / 2)
```

```r
# Visualize the skewed population

hist(
  skewed_population,
  breaks = 30,
  main = "Skewed Population",
  xlab = "Value",
  ylab = "Frequency",
  col = "lightblue"
)
```

```r
# Store sample means from the skewed population

clt_means <- c()
```

```r
# Generate many sample means

for (i in 1:1000) {
  sample_values <- sample(skewed_population, size = 30)
  sample_mean <- mean(sample_values)
  clt_means <- c(clt_means, sample_mean)
}
```

```r
# Visualize the sampling distribution

hist(
  clt_means,
  breaks = 30,
  main = "Sampling Distribution of Sample Means",
  xlab = "Sample Mean",
  ylab = "Frequency",
  col = "lightblue"
)
```

## Confidence Intervals

```r
# Take a random sample

sample_values <- sample(population, size = 50)
```

```r
# Calculate the sample mean

sample_mean <- mean(sample_values)

cat("Sample mean:", round(sample_mean, 2), "\n")
```

```r
# Calculate the margin of error

sample_std <- sd(sample_values)

margin_error <- 1.96 * (sample_std / sqrt(length(sample_values)))
```

```r
# Calculate the confidence interval

lower <- sample_mean - margin_error
upper <- sample_mean + margin_error

cat("95% Confidence Interval:\n")
cat(round(lower, 2), "to", round(upper, 2), "\n")
```
